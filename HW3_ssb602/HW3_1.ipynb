```python
from __future__ import print_function
__author__= 'ssb602'

import pylab as pl
import numpy as np
# this package sets up pretty plots
import seaborn
from scipy.optimize import curve_fit, minimize
    
%pylab inline
```

## Chi square distribution

```python
mymean = 100
```

```python
chisq_distribution = np.random.chisquare(mymean, size=100)
```

```python
pl.hist(chisq_distribution, bins=30)
pl.xlabel("x")
pl.ylabel("N")
```

## Mean and Standard deviation of chi square distribution

```python
print("Chisq mean: %.2f, Standard Deviation: %.2f"%(chisq_distribution.mean(), chisq_distribution.std()))
```

## For 100 Chi square distributions

```python
mysize = (2000 / (np.array(range(1, 100)))).astype(int)
print (mysize, mysize.shape)
```

```python
md = {} 
md['chisq'] = {}
    
for n in mysize:
    md['chisq'][n] = np.random.chisquare(mymean, size=n)

md['chisq']['means'] = {}

axchisq = pl.figure(figsize=(10,6)).add_subplot(111)

for nn in md['chisq']:
     if not type(nn) == str:
        md['chisq']['means'][nn] = md['chisq'][nn].mean()
        axchisq.plot(nn, md['chisq']['means'][nn],'o')
        axchisq.set_xlabel("Sample size", fontsize=18)
        axchisq.set_ylabel("Sample mean", fontsize=18)
        axchisq.set_title("Chi Squared", fontsize=18)
        axchisq.plot([min(mysize), max(mysize)],[mymean, mymean], 'k')
```

## Histogram of mean of samples

```python
chisq_allmeans = list(md['chisq']['means'].values())

pl.figure(figsize=(10,10))
pl.hist(chisq_allmeans, bins=30)
pl.xlabel("Sample mean")
pl.ylabel("N")
```
